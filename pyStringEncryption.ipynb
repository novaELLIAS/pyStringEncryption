### Version 1 Random substitution encryption

在这个版本中实现了针对ASCII码在33到126之间的字符的简单随机替换加密, 使用字典来构建字符映射.

在对象初始化时随机生成字符映射.

`printDic`函数主要用于打印字典.

`convertChar`函数用于对一个字符的加密和解密, `src`为源字符, `isEnc`为加密解密操作的`flag`, 为真值时加密, 反之解密.

`encryt`函数用于对整个串的加密. `src`为源串, `isEnc`定义与上文类似.

```python
# SingleReplace.py
# -*- coding: UTF-8 -*-

import random


class SingleReplaceEncryptor(object):

    dicEnc = dict()
    dicDec = dict()

    def __init__(self):

        oriAlphabet = [chr(i) for i in range(33, 127)]
        encAlphabet = oriAlphabet.copy()
        random.shuffle(encAlphabet)

        SingleReplaceEncryptor.dicEnc.clear()
        SingleReplaceEncryptor.dicDec.clear()

        for i in range(len(oriAlphabet)):
            SingleReplaceEncryptor.dicEnc[oriAlphabet[i]] = encAlphabet[i]
            SingleReplaceEncryptor.dicDec[encAlphabet[i]] = oriAlphabet[i]

    def printDic(self):
        for key in SingleReplaceEncryptor.dicEnc.keys():
            print("{} -> {}".format(key, SingleReplaceEncryptor.dicEnc[key]))

    def convertChar(self, src: str, isEnc: bool) -> str:
        if 33 <= ord(src) < 127:
            if isEnc:
                return SingleReplaceEncryptor.dicEnc[src]
            else:
                return SingleReplaceEncryptor.dicDec[src]
        else:
            return str

    def encrypt(self, src: str, isEnc: bool) -> str:
        ret = ''
        for ch in src:
            ret += SingleReplaceEncryptor.convertChar(self, ch, isEnc)
        return ret
```

### Version 2 ASCII offset encryption

简单实现了对ASCII码在33到126的字符的ASCII码偏移加密. 初始化时给`offset`赋值. `convertChar`与`encrypt`函数定义及作用与上文类似.

```python
# ASCIIoffset.py
# -*- coding: UTF-8 -*-


class ASCIIoffsetEncryptor:

    offset = 0

    def __init__(self, offset: int):
        ASCIIoffsetEncryptor.offset = offset % 94

    def convertChar(self, src: str, flag: bool) -> str:
        if 33 <= ord(src) <= 126:
            if flag:
                return chr(33 + (ord(src) - 33 + ASCIIoffsetEncryptor.offset) % 94)
            else:
                return chr(33 + (ord(src) - 33 - ASCIIoffsetEncryptor.offset + 94) % 94)
        else:
            return src

    def encrypt(self, src: str, isEnc: bool) -> str:
        ret = ''
        for ch in src:
            ret += ASCIIoffsetEncryptor.convertChar(self, ch, isEnc)
        return ret
```

### Version 3 Key encryption (token encryption)

简单实现了对小写英文字母的密钥短语加密.

`utility`模块中`generate_random_str`生成随机小写英文字符列表, `remove_same_str`用来去掉相同串中的重复元素, 用于对`token`的处理.

初始化时会选择是输入密钥和定位点或随机生成.

`pintToken`函数用于输出当前`token`以及定位点.

`convertChar`与`encrypt`函数定义及作用与上文类似.

```python
# utility.py
# -*- coding: UTF-8 -*-

import random as random


def generate_random_str(len: int) -> list:
    ret = list()
    for i in range(len):
        ret.append(chr(random.randint(ord('a'), ord('z'))))
    return ret


def remove_same_str(src: list) -> list:
    ret = list()
    for i in src:
        if i not in ret:
            ret.append(i)
    return ret
```

```python
# tokenCrypt.py
# -*- coding: UTF-8 -*-

import utility as util
import random


class TokenEncryptor:
    token = list()
    startPos = 0
    dicEnc = dict()
    dicDec = dict()

    def __init__(self):
        opt = input("Generate or input token? (G/else): ")
        if opt == "G":
            opt = int(input("Len: "))
            TokenEncryptor.token = util.generate_random_str(opt)
            TokenEncryptor.startPos = random.randint(0, 25)
            print("token: ", end="")
            print(TokenEncryptor.token)
            print("startPos: ", end="")
            print(TokenEncryptor.startPos)
        else:
            TokenEncryptor.token = list(input("Please input your token: "))
            TokenEncryptor.startPos = int(input("Please input your startPos: "))
        TokenEncryptor.token = util.remove_same_str(TokenEncryptor.token)
        dicEncTmp = list()
        for i in TokenEncryptor.token:
            dicEncTmp.append(ord(i))
        for i in range(ord("a"), ord("z") + 1):
            if chr(i) not in TokenEncryptor.token:
                dicEncTmp.append(i)
        for i in range(len(dicEncTmp)):
            TokenEncryptor.dicEnc[chr(ord("a") + (i + TokenEncryptor.startPos) % 26)] = chr(dicEncTmp[i])
            TokenEncryptor.dicDec[chr(dicEncTmp[i])] = chr(ord("a") + (i + TokenEncryptor.startPos) % 26)

    def printToken(self):
        print("token: ", end="")
        print(TokenEncryptor.token)
        print("startPos: ", end="")
        print(TokenEncryptor.startPos)

    def convertChar(self, src: str, isEnc: bool) -> str:
        if ord("a") <= ord(src) <= ord("z"):
            if isEnc:
                return TokenEncryptor.dicEnc[src]
            else:
                return TokenEncryptor.dicDec[src]
        else:
            return src

    def encrypt(self, src: str, isEnc: bool) -> str:
        ret = ""
        for ch in src:
            ret += TokenEncryptor.convertChar(self, ch, isEnc)
        return ret
```
